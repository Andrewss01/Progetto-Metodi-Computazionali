```python
df.index = df["Time"]
cross_tab = pd.crosstab(index = df.index.year, columns = df.index.month)
cross_tab
```

```python
tri_1  = cross_tab.loc[:,0:3].sum(axis =1)
tri_2  = cross_tab.loc[:,3:6].sum(axis =1)
tri_3  = cross_tab.loc[:,6:9].sum(axis =1)
tri_4  = cross_tab.loc[:,9:12].sum(axis =1)

```

```python
fig,ax  = plt.subplots(2,2,figsize = (16,10))
ax[0,0].scatter(cross_tab.index, tri_1)
ax[1,0].scatter(cross_tab.index, tri_2)
ax[0,1].scatter(cross_tab.index, tri_3)
ax[1,1].scatter(cross_tab.index, tri_4)
ax[0,0].set_title("Trimestre_1")
ax[1,0].set_title("Trimestre_2")
ax[0,1].set_title("Trimestre_3")
ax[1,1].set_title("Trimestre_4")
fig.suptitle('Numero di terremoti vs Anni', size = 15)
plt.show()
```

```python
X =[]
Y = []
contatore = 0

for i in range(2001,2020,1):
    if i in cross_tab.index:
        for j in[0,3,6,9]:
            tri_j  = cross_tab.loc[i,j:j+3].sum()
            contatore += 1
            Y.append(tri_j)
            X.append(contatore)
    else:
        for j in[0,3,6,9]:
            tri_j  = 0
            contatore += 1
            Y.append(tri_j)
            X.append(contatore)

        
plt.figure(figsize = (16,9))
plt.scatter(X,Y)
plt.xlabel("Trimestri 2001-2020")
plt.ylabel("Numero di terremoti")
plt.title("Numero di terremoti vs trimestri")
plt.show()
print(len(Y))
```

```python
X1 =[]
Y1 = []
err2 = []
contatore = 0
for i in range(2015,2020,1):
    if i in cross_tab.index:
        for j in[0,3,6,9]:
            tri_j  = cross_tab.loc[i,j:j+3].sum()
            contatore += 1
            err2.append(np.sqrt(tri_j))
            Y1.append(tri_j)
            X1.append(contatore)
    else:
        for j in[0,3,6,9]:
            tri_j  = 0
            contatore += 1
            err2.append(np.sqrt(tri_j))
            Y1.append(tri_j)
            X1.append(contatore)

        
plt.figure(figsize = (16,9))
plt.scatter(X1,Y1, color = 'red')
plt.xlabel("Trimestri 2015-2020")
plt.ylabel("Numero di terremoti")
plt.title("Numero di terremoti vs trimestri")
plt.show()
print(len(Y1))
print(len(err2))
```

```python
df2.index = df2["Time"]
X2 =[]
Y2 =[]
contatore = 0

for j in np.arange(2015,2020,1):
    a = df2[df2.index.year == j][["Sollevamento", "Time"]]
    a.index = a["Time"]
    for i in (0,3,6,9):
        z1=pd.to_numeric(a[a.index.month == i]["Sollevamento"])
        z2=pd.to_numeric(a[a.index.month == i]["Sollevamento"])
        z3=pd.to_numeric(a[a.index.month == i]["Sollevamento"])

        mean = z1.mean()+z2.mean()+z3.mean()
        contatore += 1
        Y2.append(mean)
        X2.append(contatore)
    
plt.figure(figsize = (16,9))
plt.scatter(X2,Y2)
plt.xlabel("Trimestri 2015-2019")
plt.ylabel("Sollevamento")
plt.title("Sollevamento vs Trimestri")
plt.show()
print(len(Y2))
```

```python
df2.index = df2["Time"]
x2 =[]
y2 =[]
contatore = 0

for j in np.arange(2015,2020,1):
    a = df2[df2.index.year == j][["Sollevamento", "Time"]]
    a.index = a["Time"]
    for i in range(1,13):
        z1=pd.to_numeric(a[a.index.month == i]["Sollevamento"])
        mean = z1.mean()
        contatore += 1
        y2.append(mean)
        x2.append(contatore)
    
plt.figure(figsize = (16,9))
plt.scatter(x2,y2)
plt.xlabel("Mesi 2015-2019")
plt.ylabel("Sollevamento")
plt.title("Sollevamento vs Mesi ")
plt.show()
print(len(y2))
```

```python
from iminuit import Minuit
from iminuit.cost import LeastSquares
def line(x,a,b):
    return a*x+b

Y_2 = np.array(y2)
Y_1 =  np.array(y1)
least_squares = LeastSquares(Y_2, Y_1,error, model = line)
mline = Minuit(least_squares, a = 128,b = -27)

mline.migrad()
```

```python
plt.figure(figsize=(9,6), dpi=100)
plt.errorbar(Y_2,Y_1,error, fmt = "o", label="data")
plt.plot(Y_2, line(Y_2, *mline.values), label="fit")
plt.ylabel('numero di terromoti')
plt.xlabel('Sollevamento 2015-2020')
plt.show()
```

```python
import matplotlib.gridspec as gridspec
plt.figure(figsize=(9,6),dpi=100)
gs = gridspec.GridSpec(8, 1)
ax0 = plt.subplot(gs[:5, :])
ax1 = plt.subplot(gs[6:, :])
ax0.errorbar(Y_2, Y_1, error, fmt="o")
ax0.plot(Y_2, Y_1, label="data")
ax0.plot(Y_2, line(Y_2, *mline.values), label="fit")

ax0.set_xlabel('x')
ax0.set_ylabel('$\Delta N/\Delta x$')
ax1.errorbar(Y_2, Y_1-line(Y_2, *mline.values),error, fmt="o")
ax1.plot([Y_2[0],Y_2[-1]], [0,0])
ax1.set_xlabel('x')
ax1.set_ylabel('res.')
ax0.set_ylim(0)
plt.show()
```

```python
def poly_line(x, par):
    return np.polyval(par, x)
```

```python
least_squares_np = LeastSquares(Y_2,Y_1, error, poly_line)
mpol= Minuit(least_squares_np, (600,-320,51), name=("a", "b", "c"))
mpol.migrad()
```

```python
plt.figure(figsize=(9,6), dpi=100)
plt.errorbar(Y_2,Y_1,error, fmt = "o", label="data")
plt.plot(Y_2, poly_line(Y_2, mpol.values), label="fit")
plt.ylabel('numero di terromoti')
plt.xlabel('Sollevamento 2015-2020')
plt.show()
```

```python
import matplotlib.gridspec as gridspec
plt.figure(figsize=(9,6),dpi=100)
gs = gridspec.GridSpec(8, 1)
ax0 = plt.subplot(gs[:5, :])
ax1 = plt.subplot(gs[6:, :])
ax0.errorbar(Y_2, Y_1, error, fmt="o")
ax0.plot(Y_2, Y_1, label="data")
ax0.plot(Y_2, poly_line(Y_2, mpol.values), label="fit")

ax0.set_xlabel('x')
ax0.set_ylabel('$\Delta N/\Delta x$')
ax1.errorbar(Y_2, Y_1-poly_line(Y_2, mpol.values),error, fmt="o")
ax1.plot([Y_2[0],Y_2[-1]], [0,0])
ax1.set_xlabel('x')
ax1.set_ylabel('res.')
ax0.set_ylim(0)
plt.show()
```

```python
def exp(x,a,b):
    return np.exp(-x*b)*a
least_squares_np = LeastSquares(Y_2,Y_1, error, exp)
mexp= Minuit(least_squares_np, a = 1.22,b = -7, name=("a", "b"))
mexp.migrad()
```

```python
plt.figure(figsize=(9,6), dpi=100)
plt.errorbar(Y_2,Y_1,error, fmt = "o", label="data")
plt.plot(Y_2, exp(Y_2, *mexp.values), label="fit")
plt.ylabel('numero di terromoti')
plt.xlabel('Sollevamento 2015-2020')
plt.show()
```
